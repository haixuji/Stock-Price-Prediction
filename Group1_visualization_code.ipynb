```python
import pandas as pd
aapldf = pd.read_csv('AAPL.csv')
aapldf.head()
```

```python
import pandas as pd
aapldf = aapldf.fillna(method = 'ffill')
aapldf['Date'] = pd.to_datetime(aapldf.Date)
aapldf.head().style.set_properties(subset=['Date','Close'],**{'background-color':'yellow'})
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure(figsize = (30,20))
plt.subplot(2,2,1)
plt.plot(aapldf['Date'],aapldf['Close'],color='blue')
plt.title('Apple Stock Price From 1980-2021')
```

```python
import matplotlib.pyplot as plt 
year2020 = aapldf[aapldf.Date.between('2020-01','2021-01')]
fig = plt.figure(figsize = (15,10))
plt.subplot(4,1,1)
plt.plot(year2020['Date'],year2020['Close'],color='red')
fig.suptitle('In 2020 Apple Stock close price')
```

```python
import matplotlib.pyplot as plt 
fig = plt.figure(figsize = (15,10))
plt.subplot(4,1,1)
plt.plot(year2020['Date'],year2020['Volume'],color='green')
fig.suptitle('In 2020 Apple Stock volume')
```

```python
fig = plt.figure(figsize=(15,15))
fig.tight_layout()
plt.subplot(4, 1, 1)
plt.plot(year2020['Date'], year2020['Close'])
plt.plot(year2020['Date'], year2020['Open'])
plt.legend(["Close", "Open"])
fig.suptitle('In 2020 comparision of Apple Stock close and open price')
```

```python
fig = plt.figure(figsize=(15,15))
plt.subplot(2, 2, 1)
plt.plot(aapldf['Date'], aapldf['Close'].rolling(50).mean(),color='blue')
plt.plot(aapldf['Date'], aapldf['Close'].rolling(200).mean(),color='red')
plt.legend(["50 day average", "200 day average"])
plt.title('Apple Close Price moving average')
```

```python
import seaborn as sns 
fig = plt.figure(figsize = (15,12))
fig.tight_layout()
plt.subplot(4, 1, 1)
sns.histplot(aapldf['Close'],color='darkred', kde=True)
plt.axvline(aapldf['Close'].mean(), color='k', linestyle='dashed', linewidth=2)
```

```python
aapl_numeric = aapldf.select_dtypes(exclude = ["bool"])
aapl_numeric.hist(figsize=(18,12))
plt.show()
```
